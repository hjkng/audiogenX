```python
import torch
import torchaudio
from audiocraft.utils.notebook import display_audio
```

```python
from audiocraft.models import Explainer
from audiocraft.models import AudioGen

audiogen = AudioGen.get_pretrained('facebook/audiogen-medium')
explainer = Explainer(audiogen, audiogen.lm)
```

```python
propmts = ['a_clock_ticktocks_with_music',
           'a_crowd_is_clapping_at_an_animal_of_some_kind',
           'a_train_horn_honks_as_a_train_speeds_past',
           'a_vehicle_accelerating_in_the_distance_then_drving_by']
```

```python
idx = 0 # 0 ~ 3
```

```python
original_audio = torchaudio.load(f'./{propmts[idx]}/original.wav')
audiogenX_audio = torchaudio.load(f'./{propmts[idx]}/AudioGenX.wav')
mask = torch.load(f'./{propmts[idx]}/mask.pt')
```

```python
prompt = ' '.join(propmts[idx].split('_'))
```

```python
prompt
```

```python
start = 0
end = 254
```

```python
for tkn, m in zip(explainer.lm.condition_provider.conditioners.description.t5_tokenizer.tokenize(prompt), mask[start:end].mean(dim=0)):
    print(f'{tkn} : {m:.2f}')
print(f'Size : {mask.mean().item():.2f}')
```
